```python
dummies = pd.get_dummies(df[['Car Model']]).astype(int)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final = merged.drop(['Car Model','Car Model_Mercedez Benz C class'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
X = final[['Mileage','Age(yrs)','Car Model_Audi A5','Car Model_BMW X5']]
X
```

```python
y = final[['Sell Price($)']]
y
```

```python
lr.fit(X, y)
```

```python
lr.predict([[45000, 4, 0, 0]]) #Price of mercedez benz that is 4 yr old with mileage 45000
```
